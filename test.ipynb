```python
import matplotlib.pyplot as plt
from Environment.Environment import get_environment
import torch

# Initialize the environment with a specific game
env_name = "ALE/Assault-v5"  # Example environment
env = get_environment(env_name)

# Reset the environment to get the initial state
state = env.reset()
print(state)
plt.imshow(state['pixels_trsf'].permute(1, 2, 0), cmap='gray')
plt.title("Initial State")
plt.show()

# Display the initial state image after transformations

for t in range(10):
    state['action'] = torch.zeros(2)
    state['action'][0] = 1

    plt.imshow(state['pixels_trsf'].permute(1, 2, 0), cmap='gray')
    plt.title("Initial State")
    plt.show()

    state = env.step(state)
    print(state)



# # Display the next state image after transformations
# plt.imshow(next_state['pixels_trsf'][0], cmap='gray')
# plt.title("Next State after Random Action")
# plt.show()

# print("Reward:", reward)
# print("Done:", done)
```
